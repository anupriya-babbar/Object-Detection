```python
#'D:\\KJ\\Nagesh\\Downloads\\face_recognition\\haarcascade_frontalface_default.xml'
face_cascade = cv2.CascadeClassifier('C:\\Users\\Anupriya\\Documents\\GitHub\\image\\face recog\\FaceRecognition-master\\HaarCascade\\haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('C:\\Users\\Anupriya\\Anaconda3\\Lib\\site-packages\\cv2\\data\\haarcascade_eye.xml')
```

```python
#face_cascade = cv2.CascadeClassifier('opencv-files/haarcascade_frontalface_default.xml')
```

```python
#haar_cascade_face = cv2.CascadeClassifier('OpenCV\opencv\data\haarcascades\haarcascade_frontalface_default.xml')
```

```python
faces_rects = face_cascade.detectMultiScale(gray1, scaleFactor = 1.2, minNeighbors = 5)

# Let us print the no. of faces found
print('Faces found: ', len(faces_rects))### take picture multi faces
```

```python
# Here is a list of the most common parameters of the detectMultiScale function : 
# scaleFactor :
# Parameter specifying how much the image size is reduced at each image scale. 
# minNeighbors Parameter specifying how many neighbors each candidate rectangle should have to retain it. 
# minSize : Minimum possible object size
```

```python
faces = face_cascade.detectMultiScale(gray1, 1.005, 5)
for (x,y,w,h) in faces:
    img = cv2.rectangle(gray1,(x,y),(x+w,y+h),(255,0,0),2)
    roi_gray = gray1[y:y+h, x:x+w]
    roi_color = gray1[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_gray)
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)

cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import sys

cascPath = sys.argv[1]
faceCascade = cv2.CascadeClassifier(cascPath)
```

```python
video_capture = cv2.VideoCapture(0)
while True:
    # Capture frame-by-frame
    ret, frame = video_capture.read()
```

```python
 # Capture frame-by-frame
    ret, frame = video_capture.read()

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    faces = face_cascade.detectMultiScale(
        gray,
        scaleFactor=1.1,
        minNeighbors=5,
        minSize=(30, 30),
        flags=cv2.cv.CV_HAAR_SCALE_IMAGE
    )

    # Draw a rectangle around the faces
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)

    # Display the resulting frame
    cv2.imshow('Video', frame)
```

```python
if cv2.waitKey(1) & 0xFF == ord('q'):
    break
```

```python
video_capture.release()
cv2.destroyAllWindows()
```
